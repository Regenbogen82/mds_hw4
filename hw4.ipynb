**Уровень 0**

посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
def diff_first(x):
    return (5 + 4 * 2 * np.cos(2 * x) + 3 * -np.sin(3 * x + 3))
def diff_second(x):
    return (-16 * np.sin(2 * x) + 9 * -np.cos(3 * x + 3))
```

```python
print('первая производная в точке х=1 равна:', diff_first(1), 'вторая производная в точке х=1 равна:', diff_second(1))
```

**Уровень 1**

Постройте график первой и второй производной функции
(sin(2x+1))^^2
на отрезке [-5; 5]

```python
def func(x):
    return (np.sin(2 * x + 1) ** 2)
def first_diff(x):
    return (4 * np.cos(2 * x + 1) * np.sin(2 * x + 1))
def second_diff(x):
    return (8 * np.cos(4 * x + 2))
```

```python
y = []
y1 = []
y2 = []
x = []
for i in np.linspace(-5, 6, 1000):
    x.append(i)
    y.append(func(i))
    y1.append(first_diff(i))
    y2.append(second_diff(i))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[18, 10])
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['функция', 'первая производная', 'вторая производная'])
plt.show()
```

```python
def f(x):
    return (x ** 3)
def f2(x):
    return (3 * (x**2))
```

```python
y3 = []
y4 = []
x2 = []
for i in np.linspace(-10, 11, 1000):
    x2.append(i)
    y3.append(f(i))
    y4.append(f2(i))
    
```

```python
plt.figure(figsize=[10, 10])
plt.plot(x2, y3)
plt.plot(x2, y4)
plt.plot([0, 0], [-1000, 1500])
plt.plot([-50, 50], [0, 0])
plt.legend(['функция', 'первая производная'])
plt.show()
```
